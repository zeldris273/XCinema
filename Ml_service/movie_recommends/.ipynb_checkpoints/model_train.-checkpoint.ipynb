```python
#Content-Based Filtering
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies_tv.csv")
```

```python
df["content"] = (
    df["Title"].fillna('') + " " +
    df["Overview"].fillna('') + " " +
    df["Genres"].fillna('') + " " +
    df["Director"].fillna('') + " " +
    df["Actors"].fillna('')
)
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(df["content"])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend(title, top_n=5):
    # Tìm index của phim cần gợi ý
    idx = df[df["Title"].str.lower() == title.lower()].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]  # bỏ qua chính nó
    movie_indices = [i[0] for i in sim_scores]
    return df.iloc[movie_indices][["Id", "Title", "Type"]]
```

```python
print(recommend("Solo Leveling", top_n=8))
```

```python
#Collaborative Filtering
```

```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
import pickle
```

```python
df = pd.read_csv("ratings.csv")
```

```python
df = df[['UserId', 'MediaId', 'RatingValue']]
```

```python
reader = Reader(rating_scale=(1, 10))  # thang điểm 1–10
data = Dataset.load_from_df(df, reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
model = SVD()
model.fit(trainset)
```

```python
predictions = model.test(testset)
print("RMSE:", accuracy.rmse(predictions))
```

```python
with open("recommend_model.pkl", "wb") as f:
    pickle.dump(model, f)

print("✅ Model saved to recommend_model.pkl")
```

```python
#Dự đoán cho 1 user
```

```python
with open("recommend_model.pkl", "rb") as f:
    model = pickle.load(f)
```

```python
# Gợi ý top phim cho user 1
all_movie_ids = df['MediaId'].unique()
user_id = 1

predictions = [(mid, model.predict(user_id, mid).est) for mid in all_movie_ids]
recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:10]

print("🎬 Top 10 gợi ý cho User", user_id)
for mid, score in recommendations:
    print(f"Movie {mid} → Dự đoán rating: {score:.2f}")
```
